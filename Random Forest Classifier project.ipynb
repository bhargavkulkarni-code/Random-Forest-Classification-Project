```python
numerical=[fea for fea in real_estate.columns if real_estate[fea].dtypes!=object]
numerical
```

```python
categorical=[fea for fea in real_estate.columns if real_estate[fea].dtypes==object]
categorical
```

```python
numerical=real_estate[numerical]
numerical
```

```python
categorical=real_estate[categorical]
categorical
```

```python
propertynumerical= numerical.fillna(numerical.mean())
propertynumerical
```

```python
propertycategorical=categorical.fillna(categorical.mode().iloc[0])
propertycategorical
```

```python
propertynumerical.isnull().sum()
```

```python
propertycategorical.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
propertycategorical['Property']=le.fit_transform(propertycategorical['Property'])
```

```python
list(le.classes_)
```

```python
x=propertynumerical[['num_rooms','property_tax_rate']]
x
```

```python
y=propertycategorical['Property']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.6,random_state=12)
```

```python
from collections import Counter
Counter(y_train)
```

# This is Balanced data

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rm= RandomForestClassifier()
```

```python
rm.fit(x_train,y_train)
```

```python
y_prediction=rm.predict(x_test)
y_prediction
```

```python
rm.score(x_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(y_test,y_prediction)
cm
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.heatmap(cm,annot=True,fmt='.2f',xticklabels=['?', 'Four Family', 'Single Family', 'Three Family', 'Two Family'],yticklabels=['?', 'Four Family', 'Single Family', 'Three Family', 'Two Family'])
plt.xlabel('Predicted class')
plt.ylabel('Actual class')
```
